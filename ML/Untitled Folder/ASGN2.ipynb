### Classify the email using the binary classification method. Email Spam detection has two states: 
a) Normal State Not Spam

b) Abnormal State Spam. Use K-Nearest Neighbors 

and Support Vector Machine for classification. Analyze their performance. 
Dataset link: The emails.csv dataset on the Kaggle 
https://www.kaggle.com/datasets/balaka18/email-spam-classification-dataset-csv

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Tanmay49/SEM7/main/ML/emails.csv')
df.head()
```

```python
print(df['Prediction'].unique())
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
X = df.iloc[:, 1:3001]
Y = df.iloc[:, -1]  ## prediction column
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
# x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

x_train,x_test,y_train,y_test=train_test_split(X,Y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC ## note syntax
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

# Analysis of the Classification Reports:
K-Nearest Neighbors (KNN):

High precision and recall for Class 0.
Class 1 has decent recall and slightly lower precision.
Overall, the accuracy is around 90%, showing that the model performs well for this dataset.
Support Vector Classifier (SVC):

Good precision for both classes, but recall for Class 1 is significantly lower (0.43).
This implies that the SVC struggles to correctly identify instances of Class 1, leading to a lower F1-score and poorer balance overall.
The accuracy is lower (82%), and the macro average F1-score is 0.73, indicating poorer performance, especially for Class 1.
Conclusion:
Ensure the training (model.fit) is only done on the training set (x_train, y_train) to maintain the integrity of the model evaluation.
The KNN model seems to perform better than the SVC for this dataset, as indicated by the higher F1-scores and better recall for Class 1.
